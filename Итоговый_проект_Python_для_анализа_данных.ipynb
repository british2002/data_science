Рассчитайте основные статистики для переменных (среднее,медиана,мода,мин/макс,сред.отклонение).


```python
#выделим столбцы с категориальными данными
w_accident = df['Work_accident']
left = df['left']
promo_l_5 = df['promotion_last_5years']
department = df['department']
salary = df['salary']
```

```python
#для столбца last_evaluation рассчитаем базовые статистики: среднее, медиану, моду, максимальную и минимальную величины, среднее отклонение
l_e_mean = last_evaluation.mean()
l_e_median = last_evaluation.median()
l_e_mode = last_evaluation.mode()[0]
l_e_min = last_evaluation.min()
l_e_max = last_evaluation.max()
l_e_std = last_evaluation.std()

#выведем результат
print('Расчет базовых статистик по столбцу last_evaluation:')
print(f'Среднеарифметическое: {l_e_mean}')
print(f'Медиана: {l_e_median}')
print(f'Мода: {l_e_mode}')
print(f'Минимальное значения: {l_e_min}')
print(f'Максимальное значения: {l_e_max}')
print(f'Среднее отклонение: {l_e_std}')
```

# 8
Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):


*   Доля сотрудников с повышением за последние 5 лет
*   Средняя степень удовлетворенности
*   Среднее количество проектов

```python
#отфильтруем уволенных и получивших повышение сотрудников и запишем в переменную; для удобства оставим только 1 столбец, например, зарплаты
df_left_promo = int(df.loc[(df['left'] == 1) & (df['promotion_last_5years'] == 1)].count()[['salary']])
df_left = int(df.loc[(df['left'] == 1)].count()[['salary']])

left_promo_perc = round((df_left_promo * 100 / df_left),2)

#отфильтруем оставшихся и получивших повышение сотрудников и запишем в переменную; для удобства оставим только 1 столбец, например, зарплаты
df_promo = int(df.loc[(df['left'] == 0) & (df['promotion_last_5years'] == 1)].count()[['salary']])
df_stay = int(df.loc[(df['left'] == 0)].count()[['salary']])

stay_promo_perc = round((df_promo * 100 / df_stay),2)

print(f'Процент (доля к 100%) сотрудников с повышением в последние 5 лет среди уволившихся: {left_promo_perc} %')
print(f'Процент (доля к 100%) сотрудников с повышением в последние 5 лет среди оставшихся: {stay_promo_perc} %')
```

```python
#отфильтруем уволенных сотрудников и расчитаем среднее арифметическое по столбцу satisfaction_level
df_left_satisfaction = round(float(df.loc[(df['left'] == 1)].mean()[['satisfaction_level']]),2)

#отфильтруем оставшихся сотрудников и расчитаем среднее арифметическое по столбцу satisfaction_level
df_satisfaction = round(float(df.loc[(df['left'] == 0)].mean()[['satisfaction_level']]),2)

print(f'Средняя степень удовлетворенности среди уволившихся: {df_left_satisfaction}')
print(f'Средняя степень удовлетворенности среди оставшихся: {df_satisfaction}')
```

```python
#отфильтруем уволенных сотрудников и расчитаем среднее арифметическое по столбцу number_project
df_left_projects = round(float(df.loc[(df['left'] == 1)].mean()[['number_project']]),2)

#отфильтруем оставшихся сотрудников и расчитаем среднее арифметическое по столбцу number_project
df_projects = round(float(df.loc[(df['left'] == 0)].mean()[['number_project']]),2)

print(f'Среднее количество проектов среди уволившихся: {df_left_projects}')
print(f'Среднее количество проектов среди оставшихся: {df_projects}')
```

# 9
Разделить данные на тестовую и обучающую выборки. Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary). Оценить качество модели на тестовой выборки



```python
#определим Х и У
X = df[['promotion_last_5years', 'Work_accident']]
y = df['left']

# разделяем выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# импортируем LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# обучаем модель
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# смотрим разницу факта и прогноза
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
```

```python
result
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, lda.predict(X_test))
```

Качество тестовой модели достаточно высокое и составляет 0.7608
